#Pipeline : Clustering basé sur la détection d’objets

###Étape 1 : Importer les librairies

```python
import os
import cv2
import numpy as np
from sklearn.cluster import KMeans
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input
from tensorflow.keras.preprocessing import image
```

###Étape 2 : Détection et extraction de features des objets

Ici, tu peux utiliser un modèle pré-entraîné (MobileNetV2 par exemple) pour extraire des features de l’image qui représentent son contenu (routes, bâtiments, etc.).

```python
# Dossier des images redimensionnées
input_folder = "/content/drive/MyDrive/master SDIA/S3/Reinforcement Learning/projet/images_redimensionnees"

# Charger MobileNetV2 pour extraire les features
model = MobileNetV2(weights='imagenet', include_top=False, pooling='avg')

features_list = []
image_names = []

for filename in os.listdir(input_folder):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        img_path = os.path.join(input_folder, filename)
        img = image.load_img(img_path, target_size=(256, 256))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array = preprocess_input(img_array)

        features = model.predict(img_array)
        features_list.append(features.flatten())
        image_names.append(filename)

features_array = np.array(features_list)
```

###Étape 3 : Clustering en 3 clusters

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(features_array)

labels = kmeans.labels_  # 0,1,2 correspondant aux clusters
```

###Étape 4 : Visualisation des résultats

```python
for img_name, label in zip(image_names, labels):
    print(f"{img_name} -> Cluster {label}")
```

```python
import pandas as pd

# Créer le DataFrame
df = pd.DataFrame({
    'commune': image_names,
    'richesse': labels  # ici 0,1,2 selon le cluster
})

# Afficher le DataFrame
df
```

```python
cluster_mapping = {0: 'High', 1: 'Medium', 2: 'Low'}
df['richesse'] = df['richesse'].map(cluster_mapping)

df
```

```python
import os
from math import ceil, sqrt
from PIL import Image

# Dossier contenant les images
input_folder = "/content/drive/MyDrive/master SDIA/S3/Reinforcement Learning/projet/images_redimensionnees"

# Liste des fichiers images
image_files = [os.path.join(input_folder, f) for f in os.listdir(input_folder)
               if f.lower().endswith((".png", ".jpg", ".jpeg"))]

# Charger toutes les images
images = [Image.open(img) for img in image_files]

if not images:
    raise ValueError("Aucune image trouvée dans le dossier !")

# Taille pour chaque image (on prend la plus petite pour éviter de trop étirer)
img_width = min(img.width for img in images)
img_height = min(img.height for img in images)

# Redimensionner toutes les images à la même taille
images_resized = [img.resize((img_width, img_height)) for img in images]

# Déterminer la taille de la grille
num_images = len(images_resized)
cols = int(ceil(sqrt(num_images)))  # colonnes approximativement carrées
rows = int(ceil(num_images / cols))  # nombre de lignes

# Créer une grande image vide pour la grille
grid_image = Image.new('RGB', (cols * img_width, rows * img_height), color=(255, 255, 255))

# Coller les images dans la grille
for idx, img in enumerate(images_resized):
    x = (idx % cols) * img_width
    y = (idx // cols) * img_height
    grid_image.paste(img, (x, y))

# Afficher l'image finale
grid_image.show()

# Optionnel : sauvegarder l'image combinée
# grid_image.save("grid_image.jpg")
```
